```python
import geopandas as gpd
import pandas as pd
import IPython.display

utm31n = 'EPSG:32631'
wgs84 = 'EPSG:4326'
```

```python
section_gdf = gpd.read_parquet('data/section.geoparquet')
sectionjunction_gdf = gpd.read_parquet('data/sectionjunction.geoparquet')
vtssector_gdf = gpd.read_parquet('data/vtssector.geoparquet')
lock_gdf = gpd.read_parquet('data/lock.geoparquet')

vtssector_gdf['buffer'] = vtssector_gdf.to_crs(utm31n).buffer(50).to_crs(wgs84)
```

```python
vtssector_gdf = vtssector_gdf.set_geometry('buffer')
html = vtssector_gdf.sjoin(lock_gdf, predicate='intersects', lsuffix='vtssector', rsuffix='lock')[['Name_lock', 'Name_vtssector']].to_html()
vtssector_gdf = vtssector_gdf.set_geometry('geometry')
IPython.display.HTML(html)
```
