```python
import pathlib
import shapely

import pandas as pd
import geopandas as gpd
```

```python
data_dir = pathlib.Path('~/data/euris').expanduser()
version = 'v0.1.0'
```

```python
excel_files = list(data_dir.glob('RisIndex*.xlsx'))
excel_files
```

```python
ris_gdfs = []
for excel_file in excel_files:
    ris_df = pd.read_excel(data_dir / excel_file)
    ris_df_geoms = gpd.points_from_xy(x=ris_df["long_"], y=ris_df["Lat"], crs="EPSG:4326")
    
    ris_gdf = gpd.GeoDataFrame(ris_df, geometry=ris_df_geoms)
    ris_gdf['path'] = excel_file.name
    ris_gdfs.append(ris_gdf)
    # ris_gdf.to_file((data_dir / excel_file).with_suffix('.gpkg'))
ris_gdf = pd.concat(ris_gdfs)
```

```python
ris_gdf.to_file(data_dir / version / f"ris_index_{version}.gpkg")
```
